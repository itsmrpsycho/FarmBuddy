```python
import os
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import random
import torch
import shutil
```

```python
train_path = "/content/drive/MyDrive/Models/disease_notebook/cropDiseasesSannidhya"
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
# model = YOLO('yolov8n-cls.yaml')  # You can use different variants of YOLOv8 (e.g., 'yolov8n.pt', 'yolov8s.pt', etc.)
# model.to(device=device)
# # Training loop

# model.train(data=train_path, epochs=1, imgsz=128)
```

```python
# Load the YOLO model
model = YOLO('yolov8n-cls.yaml')
model.to(device=device)

# Define training parameters
patience = 10  # Early stopping patience
batch_size = 16
img_size = 192
epochs = 100

# Training loop
train_results = model.train(

        data=train_path,
        epochs=epochs,
        imgsz=img_size,
        batch=batch_size,
        device=device,
        patience = patience,
        save = True,
        save_period = 5,
        project = "/content/drive/MyDrive/Models/disease_notebook/SavedModels",
        name = "model1_",
    )



print("Training complete.")
```

```python
model.val()
```

```python
random.seed(42)
def inference(testImgPath):
    img = cv2.imread(testImgPath)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = model.predict(testImgPath)
    index=results[0].probs.top1
    output=results[0].names[index]
    # print(output)
    plt.figure(figsize=(4,4))
    plt.imshow(img)
    plt.title(f"Model predicted: {output}, actual={testImgPath[testImgPath.rfind('/')+1:]}")
    plt.axis('off')
    plt.tight_layout()
    plt.show()

testFolderPath = "/content/drive/MyDrive/Models/disease_notebook/cropDiseasesSannidhya/test"
n_test_imgs = 11
imgPaths = [os.path.join(testFolderPath, filename) for filename in random.sample(os.listdir(testFolderPath), n_test_imgs)]
for imgPath in imgPaths:
    inference(imgPath)
```
